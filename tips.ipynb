It is a data set in which the values ​​associated with the total paid account in a restaurant are kept.

· total_bill -> The amount of the account paid in dollars (3.07$ — 50.81$)

· tip -> Tip amount ( 1$ — 10$)

· sex -> Gender of the person paying the bill (Male , Female)

· smoker -> Is smoking on the table? (No , Yes)

· day -> Which day people arriving (Thur,Fri,Sat,Sun)

· time -> Time of day(Lunch, Dinner)

· size -> Number of people sitting at the table (1–6)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
plt.style.use('ggplot')
# warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='time', row='smoker')
```

```python
sns.heatmap(tips.loc[:, ["total_bill", "tip", ""]].corr(), annot=True, cmap='coolwarm')
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="day", size="size", col_wrap=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="size",size="size")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```
